# <span style="color:orange">6.3.</span> 순환 신경망의 고급 사용법

+ 순환 신경망의 성능과 일반화 능력을 향상시키기 위한 세 가지 기술들을 살펴볼 것임.
    + 순환 드롭아웃(recurrent dropout): 순환 층에서 과대적합을 방지하기 위해 케라스에 내장되어 있는 드롭아웃 사용
    + 스태킹 순환 층(stacking recurrent layer): 네트워크의 표현 능력(representaional power) 증가(대신 계산 비용은 많이 증가함)
    + 양방향 순환 층(bidirectional recurrent layer): 순환 네트워크에 같은 정보를 다른 방향으로 주입하여 정확도를 높이고 기억을 좀 더 오래 유지시킴.

+ 최근 데이터(몇 일치 데이터 포인트)를 입력으로 사용하여 모델을 만들고 24시간 이후의 기온을 예측할 것임.

## <span style="color:orange">6.3.2.</span> 데이터 준비

#### 문제 정의
+ lookback 타임스텝(한 타임 스텝은 10분)만큼 이전으로 돌아가서 매 step 타임스텝마다 샘플링한 것으로 delay 타임스텝 이후의 온도를 예측할 수 있을 것인가?
    + 사용할 변수
        + lookback = 1440: 10일 번 데이터로 돌아감
        + step = 6: 1시간(1포인트는 10분, 6 스텝 = 60분)마다 데이터 포인트 하나를 샘플링
        + delay =144: 24시간이 지난 데이터가 타깃이 됨.
+ 시작하기 전에 처리해야 하는 두 가지 작업
    + 신경망에 주입할 수 있는 형태로 데이터 전처리. 각 특징점들을 정규화하여 비슷한 범위를 가진 값으로 바꿀 필요가 있음.
    + float_data 배열을 받아 과거 데이터의 배치와 미래 타깃 온도를 추출하는 파이썬 제너레이터(generator)를 만듦. 모든 샘플을 각기 메모리에 적재하는 것은 낭비가 심하기 때문에 원본 데이터를 사용하여 그때그때 배치를 만들 것임.

```python
# 데이터 정규화하기

mean = float_data[:200000].mean(axis=0)
float_data -= mean
std = float_data[:200000].std(axis=0)
float_data /= std
```

##### 제너레이터 함수
+ 이 제너레이터 함수는 (sample, targets) 튜플을 반복적으로 반환함.
    + samples는 입력 데이터로 사용할 배치
    + targets는 이에 대응하는 타깃 온도의 배열
+ 이 제너레이터 함수에는 다음과 같은 매개 변수가 있음.
    + data: 위의 코드에서 정규화한 부동 소수 데이터로 이뤄진 원본 배열
    + lookback: 입력으로 사용하기 위해 거슬러 올라갈 타임스텝
    + min_index와 max_index: 추출할 타임스텝의 범위를 지정하기 위한 data 배열의 인덱스. 검증 데이터와 테스트 데이터를 분리하는데 사용함.
    + shuffle: 샘플을 섞을지, 시간 순서대로 추출할지 결정함.
    + batch_size: 배치의 샘플 수
    + step: 데이터를 샘플링할 타임스텝 간격. 1시간에 하나의 데이터 포인트를 추출하기 위해 6으로 지정할 것임.

```python
# 시계열 데이터와 타깃을 반환하는 제너레이터 함수

def generator(data, lookback, delay, min_index, max_index,
              shuffle=False, batch_size=128, step=6):
    if max_index is None:
        max_index = len(data) - delay - 1
    i = min_index + lookback
    while 1:
        if shuffle:
            rows = np.random.randint(
                min_index + lookback, max_index, size=batch_size)
        else:
            if i + batch_size >= max_index:
                i = min_index + lookback
            rows = np.arange(i, min(i + batch_size, max_index))
            i += len(rows)

        samples = np.zeros((len(rows),
                           lookback // step,
                           data.shape[-1]))
        targets = np.zeros((len(rows),))
        for j, row in enumerate(rows):
            indices = range(rows[j] - lookback, rows[j], step)
            samples[j] = data[indices]
            targets[j] = data[rows[j] + delay][1]
        yield samples, targets
```

```python
# 훈련, 검증, 테스트 제너레이터 준비하기

lookback = 1440
step = 6
delay = 144
batch_size = 128
train_gen = generator(float_data,
                      lookback=lookback,
                      delay=delay,                   
                      min_index=0,
                      max_index=200000,
                      shuffle=True,
                      step=step,
                      batch_size=batch_size)
val_gen = generator(float_data,
                    lookback=lookback,
                    delay=delay,
                    min_index=200001,
                    max_index=300000,
                    shuffle=True,
                    step=step,
                    batch_size=batch_size)
test_gen = generator(float_data,
                     lookback=lookback,
                     delay=delay,
                     min_index=300001,
                     max_index=None,
                     shuffle=True,
                     step=step,
                     batch_size=batch_size)

val_steps = (300000-200001-lookback) // batch_size # 전체 검증 세트를 순회하기 위해 val_gen에서 추출할 횟수

test_steps = (len(float_data) - 300001 - lookback) // batch_size # 전체 테스트 세트를 순회하기 위해 test_gen에서 추출할 횟수
```

## <span style="color:orange">6.3.3.</span> 상식 수준의 기준점

+ 머신 러닝 기반의 모델이라면 **90% 이상**을 달성해야 유용하다고 볼 수 있음.
+ 이 모델에서는 일자별 온도 주기를 이용하여 기준을 설정했음.

```python
# 상식적인 기준 모델의 MAE 계산하기

def evaluate_naive_method():
    batch_maes = []
    for step in range(val_steps):
        samples, targets = next(val_gen)
        preds = samples[:, -1, 1]
        mae = np.mean(np.abs(preds - targets))
        batch_maes.append(mae)
    print(np.mean(batch_maes))
    
evaluate_naive_method() # mae 출력 - 오늘 기온과 내일 기온이 비슷하다고 가정하고 차이 분석
```

```python
# mae를 섭씨 단위로 변환하기
# 앞의 온도 데이터의 평균은 0이고 표준편차는 1(표준화를 했기 때문)이므로, 이를 섭씨 단위로 변환할 필요가 있음.
# 평균 절대값 오차 0.29에 표준 편차를 곱하여 온도계 단위의 mae를 출력.

celsius_mae = 0.29 * std[1]

print(celsius_mae)
```

## <span style="color:orange">6.3.4.</span> 기본적인 머신 러닝 방법

+ 앞에서 계산한 mae를 기준으로 하여 머신러닝의 성능을 평가할 수 있음.
+ 먼저 2개의 Dense 층을 통과시키는 완전 연결 네트워크를 만들어서 결과를 비교할 것임.

```python
# 완전 연결 모델을 훈련하고 평가하기

from keras.models import Sequential
from keras import layers
from keras.optimizers import RMSprop

model = Sequential()
model.add(layers.Flatten(input_shape=(lookback // step, float_data.shape[-1])))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1))

model.compile(optimizer=RMSprop(), loss='mae')
history = model.fit_generator(train_gen,
                              steps_per_epoch=500,
                              epochs=20,
                              validation_data=val_gen,
                              validation_steps=val_steps)
```

```python
# 결과 그래프 그리기
import matplotlib.pyplot as plt

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(loss) + 1)

plt.figure()
plt.plot(epochs, loss, 'bo', label = 'Traning loss')
plt.plot(epochs, val_loss, 'b', label = 'Validation loss')
plt.title('Traning and validation loss')
plt.legend()

plt.show()
```

+ 이 네트워크의 성능은 앞서 계산한 상식 수준의 기준점을 뛰어넘지 못함.
+ 학습 알고리즘이 특정한 종류의 간단한 모델을 찾도록 하드코딩되지 않았다면, 모델 파라미터를 학습하는 방법은 간단한 문제를 위한 간략한 해결책을 찾지 못할 수 있음.

## <span style="color:orange">6.3.5.</span> 첫 번째 순환 신경망

+ 앞선 모델은 시계열 데이터를 펼쳤기 때문에(Flatten층) 입력 데이터에서 시간 개념을 잃어버림.
+ 이번에는 인과 관계와 순서가 의미있는 시퀀스 데이터를 그대로 사용할 것임.
+ **GRU**(Gated Recurrent Unit)을 사용할 예정임.
    + GRU 층은 LSTM과 같은 원리로 작동하지만 좀 더 간결하고 계산 비용이 덜 듬
    + GRU 셀의 상태는 하나고, 하나의 게이트가 삭제 게이트와 입력 게이트의 역할을 함.
    + LSTM 만큼 표현 학습 능력이 높지는 않을 수 있음

```python
# GRU를 사용한 모델을 훈련하고 평가하기

from keras.models import Sequential
from keras import layers
from keras.optimizers import RMSprop

model = Sequential()
model.add(layers.GRU(32, input_shape=(None, float_data.shape[-1])))
model.add(layers.Dense(1))

model.compile(optimizer=RMSprop(), loss='mae')
history = model.fit_generator(train_gen,
                              steps_per_epoch=500,
                              epochs=20,
                              validation_data=val_gen,
                              validation_steps=val_steps)
```

```python
# 결과 그래프 그리기
import matplotlib.pyplot as plt

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(loss) + 1)

plt.figure()
plt.plot(epochs, loss, 'bo', label = 'Traning loss')
plt.plot(epochs, val_loss, 'b', label = 'Validation loss')
plt.title('Traning and validation loss')
plt.legend()

plt.show()
```
